```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('tested.csv')
```

```python
df.head()
```

```python
df.info()
```

```python
df.describe()
```

```python
gender = df['Sex'].value_counts()

# Plot the pie chart
gender.plot(kind='pie', autopct='%1.1f%%', startangle=90, figsize=(6, 6))

# Add a title
plt.title('Gender Distribution')

# Show the plot
plt.ylabel('')  # Remove the y-label for a cleaner look
plt.show()
```

```python
survival_rate = df.groupby('Sex')['Survived'].mean()

plt.pie(survival_rate.values, labels=survival_rate.index, autopct='%1.1f%%')

# Set title
_ = plt.title('Survival Rate by Sex')
```

```python
# @title Embarked

df.groupby('Embarked').size().plot(kind='bar', color=sns.palettes.mpl_palette('Dark2'))
plt.gca().spines[['top', 'right',]].set_visible(False)
```

```python
survival_rate = df.groupby('Embarked')['Survived'].mean()

plt.pie(survival_rate.values, labels=survival_rate.index, autopct='%1.1f%%')

# Set title
_ = plt.title('Survival Rate by Port of Embarkation')
```

```python
sns.countplot(x='Embarked', hue='Sex', data=df)
plt.xlabel('Port of Embarkation')
plt.ylabel('Count')
plt.title('Passenger Embark Distribution by Gender')
plt.show()
```

```python
survival_rate = df.groupby('Pclass')['Survived'].mean()

plt.pie(survival_rate.values, labels=survival_rate.index, autopct='%1.1f%%')

# Set title
_ = plt.title('Survival Rate by Passenger Class')
```

```python
sns.countplot(x='Pclass', hue='Sex', data=df)

# Add labels and title
plt.xlabel('Passenger Class')
plt.ylabel('Count')
plt.title('Passenger Class Distribution by Gender')

# Show the plot
plt.show()
```
